```python
from collections import Counter
from matplotlib import pyplot as plt


def plot_1_language(instr):
    with open(instr,"r") as fh:
        wordlist=[x.strip() for x in fh.readlines()]
    letters= "".join(wordlist)
    counted_letters=Counter(letters)
    for key, val in counted_letters.items():
        counted_letters[key] = val / float(len(letters))
        
    vowels =["a","e","i","o","u"]
    plt.plot (range(5),
              [counted_letters[vowel] for vowel in vowels],
              label=instr.split("\\")[-1])
    
languages=["nederlands.txt",
          "dansk.txt",
          "ukenglish.txt",
          "deutsch.txt"]


for language in languages:
    plot_1_language(language)  

plt.xticks(range(5), ["a","e","i","o","u"]) 
plt.xlabel("letter")
plt.ylabel("frequency")
plt.legend()



```
